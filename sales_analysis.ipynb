```python
from sklearn import preprocessing 
```

```python
df['norm_profit']= df['Profit']
```

```python
# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()
```

```python
# get all values of profit col
p =  df[['norm_profit']].values.astype(float)

# transform the extracted value to normal form 
prof_scaled = min_max_scaler.fit_transform(p)
```

```python
# create dataframe of normalized value 
prof_scaled_df = pd.DataFrame(prof_scaled)
```

```python
# put the normalized value back to the df
df['norm_profit'] = prof_scaled_df
```

```python
alt.Chart(df).mark_circle().encode(
    alt.X('Discount',scale=alt.Scale(zero=False)),
    alt.Y('Profit',scale=alt.Scale(zero=False,padding=1)),
    color = 'Customer Segment',
    size = 'Discount'
    
).properties(title='Discount Does Not Seems To Be Corelated With Profit')
```
